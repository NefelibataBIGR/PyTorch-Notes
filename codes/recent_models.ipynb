# 现有模型的使用、修改、保存、加载

- 以torchvision里面的 VGG16 模型为例

```python
import torchvision
import torch
from torch import nn
from torch.utils.data import DataLoader
```

```python
# 以torchvision.Dataset里面的ImageNet数据集为例
# train_data = torchvision.datasets.ImageNet("my_imagenet", split='train', transform=transforms.ToTensor())
```

*发现需要手动下载数据集，但是数据集太大了，遂放弃使用*

```python
vgg16_pre = torchvision.models.vgg16(weights=True) # weights=True表示用预训练好的参数
vgg16_unpre = torchvision.models.vgg16(weights=False) # weights=False表示用没有经过训练的参数
```

```python
vgg16_pre
```

*发现最后输出为1000类，而运用在CIFAR10数据集中需要最后为10类*

```python
train_data = torchvision.datasets.CIFAR10(root='./my_cifar10', train=True, transform=torchvision.transforms.ToTensor(), download=True)
```

- 修改模型

方法1：添加一层线性层，将输出从1000变为10

```python
vgg16_pre.add_module('add_linear', nn.Linear(1000, 10))
vgg16_pre
```

*或者vgg16_pre.classifier.add_module加到classifier的Sequential里面*

方法2：修改原有模型

```python
vgg16_unpre
```

```python
vgg16_unpre.classifier[6] = nn.Linear(4096, 10) # 注意channel值对应
vgg16_unpre
```

- 保存模型

保存整个模型

```python
torch.save(vgg16_unpre, "my_vgg16.pth") # 注意格式为 .pth
```

**只保存参数（推荐）**

```python
torch.save(vgg16_unpre.state_dict(), "my_vgg16_dict.pth") # 只保存参数，以字典形式
```

- 加载模型

整个模型

```python
model = torch.load("my_vgg16.pth", weights_only=False) # weights_only=False才不报错，表示不只加载模型的权重，还加载整个模型
print(model)
```

只加载参数

```python
model_dict = torch.load("my_vgg16_dict.pth")
print(model_dict)
```
